```python
import pandas as pd
from pandas.api.types import is_numeric_dtype
```

```python
df = pd.read_csv("videosUS.csv")
df.head()
```

```python
print(df.isna().sum()) # missing values for each variable 
print(f'Total : {df.isna().sum().sum()}')
```

```python
df.loc[:, df.isnull().any()].columns
```

```python
columns = df.columns
stat_result = {}
for col in columns : 
    if is_numeric_dtype(df[col]) :
        
        stat_result[col] = {'median':df[col].median(),'mean':df[col].mean(),'quar':df[col].quantile(q=0.25),'std':df[col].std()}
stat_result
```

```python
df['views_Z'] = (df['views']-stat_result['views']['mean'])/stat_result['views']['std']
df = df[df['views_Z']>3]
```

```python
df
```

```python
for col in columns :
    if is_numeric_dtype(df[col]) :
        df[col+"_Z"] = (df[col]-stat_result[col]['mean'])/stat_result[col]['std']
        df = df[df[col+"_Z"]>3]
```

```python
y = pd.get_dummies(df['tags'], preﬁx='tags')
y
```

```python
tags = {}
for tag in df['tags'] :
    if tags.get(tag) :
        tags[tag]+=1
    else :
        tags[tag] = 1
```
